```python
import pandas as pd
import numpy as np
```

```python
def sign_test(x):
    x1, x2 = x
    if x1 == x2:
        return None
    return '+' if x1 > x2 else '-'
```

```python
results = pd.read_excel('joined-results.xlsx', sheet_name=None)
```

```python
agg_results = []
for _, df in results.items():
    gb_df = df.groupby(['Dataset', 'Algorithm']).median()['PR-Curve']
    agg_results.append(gb_df.unstack())
    
agg_df = pd.concat(agg_results)
agg_df['Sign'] = agg_df[['Baseline', 'RandAugment']].apply(sign_test, axis=1)
```
